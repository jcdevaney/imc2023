```python
# import libraries
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display
```

```python
# Example of plotting audio function (Lab Assignment 1)
def plotAudio(filename,sampleRate):
    # Open the audio file with librosa.load()
    sig , sr = librosa.load(filename,mono=True,sr=sampleRate)
    
    # Adjust the x-axis to represent time
    time=np.arange(0,sig.shape[0])/sr
    
    # Scale the y-axis to represent amplitude from -1 to 1
    excursion=max(abs(sig))
    sigNew = sig / excursion
    
    # Label the x and y axes with variables
    xlabel = 'Time (seconds)'
    ylabel = 'Amplitude'

    # Plot the signal using the x and y axis variables    
    plt.plot(time,sigNew)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
plotAudio('avm.wav',44100)
```

```python
plotAudio('avm.wav',441)
```
