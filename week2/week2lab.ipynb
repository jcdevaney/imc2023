```python
#import libraries
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display

#import audio files
!git clone https://github.com/jcdevaney/imc2023.git
```

```python
# open the audio file avm2.wav with librosa.load()
file = 'imc2023/audioFiles/avm2.wav'
samplingRate = 44100
sig, sr = librosa.load(file,mono=True,sr=samplingRate)
```

```python
# plot time domain representation of the signal
plt.plot(sig)
```

```python
#adjust the x-axis to represent time
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)
```

```python
# scale the y-axis to represent amplitude
excursion=max(abs(sig))
sigNew = sig / excursion
plt.plot(time,sigNew)
```

```python
# add x and y labels
plt.plot(time,sigNew)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
```
